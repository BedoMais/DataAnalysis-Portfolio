### Change dataset column names

```python
#Change dataset column names
dts.rename(columns = {'ID_Pedido' : "Order_ID"}, inplace = True)
dts.rename(columns = {'Data_Pedido' : 'Order_Date'}, inplace = True)
dts.rename(columns = {'ID_Cliente' : 'Customer_ID'}, inplace = True)
dts.rename(columns = {'Segmento' : 'Segment'}, inplace = True)
dts.rename(columns = {'Pais' : 'Country'}, inplace = True)
dts.rename(columns = {'Regiao' : 'Region'}, inplace = True)
dts.rename(columns = {'ID_Produto' : 'Product_ID'}, inplace = True)
dts.rename(columns = {'Categoria' : 'Category'}, inplace = True)
dts.rename(columns = {'Nome_Produto' : 'Product_Name'}, inplace = True)
dts.rename(columns = {'Valor_Venda' : 'Sale_Price'}, inplace = True)
dts.rename(columns = {'Quantidade' : 'Amount'}, inplace = True)
```

### Logical Operators for Data Manipulation with Pandas

### Grouping Data in DataFrames with Group By

```python
#Apply groupby function
dts[['Segment','Region', 'Sale_Price']].groupby(['Segment', 'Region']).mean()
```

### Agregação Múltipla com Group By

```python
#Apply groupby function
dts[['Segment', 'Region', 'Sale_Price']].groupby(['Segment', 'Region']).agg(['mean', 'std', 'count'])
```

```python
#Imported the iris dataset from Scikit-Learn
from sklearn.datasets import load_iris
data = load_iris()
```

```python
#Load the iris dataset as a DataFrame in pandas
import pandas as pd
df = pd.DataFrame(data['data'], columns = data['feature_names'])
df['species'] = data['target']
df.head()
```

```python
#Create the graphic
df.plot()
```

```python
#Create scatter plot
df.plot.scatter(x = 'sepal length (cm)', y = 'sepal width (cm)')
```

```python
columns = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
df[columns].plot.area()
```

```python
df.groupby('species').mean().plot.bar()
```

```python
#Pizza Graphic
df.groupby('species').count().plot.pie(y = 'sepal length (cm)')
```

```python
#Graohic KDE(Kenal Density Function)
df.plot.kde(subplots = True, figsize = (8,8))
```

```python
#Boxplot
columns = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
df[columns].plot.box(figsize = (8,8))
```
