### 1- City with the highest sales volume in the "office supplies" category?

```python
#Filter the dataframe by categorizing "Office Supplies"
categorize = dataset[dataset["Category"] == "Office Supplies"]
```

```python
categorize.head()
```

```python
#Group by city and calculate total of Valor_venda
total_sales_volume = categorize.groupby("City")["Sale_Price"].sum()
```

```python
#City with the highest sales volume
City_Highest_Sales_Volume = total_sales_volume.idxmax()
print("The city with the highest sales volume in the office supplies category: ", City_Highest_Sales_Volume)
```

```python
#Checking the result
checking_result = total_sales_volume.sort_values(ascending = False)
print(checking_result)
```

### 2 Total sale by order date?

```python
#Group by Data_Pedido of Valor_venda
total_sale_by_Order_Date = dataset.groupby("Order_Date")["Sale_Price"].sum()
print(total_sale_by_Order_Date)
```

```python
#Data_Pedido with highest sale volume
Order_Date_highest_sale = total_sale_by_Order_Date.idxmax()
print(Order_Date_highest_sale)
```

### 5 Which segment had the highest total sales?

```python
#Categorizing
categorizing_by_segment = dataset.groupby('Segment')['Sale_Price'].sum().reset_index()
```

```python
result_categorizing_by_segment = categorizing_by_segment.sort_values(by = 'Sale_Price', ascending = False)
```

```python
result_categorizing_by_segment
```

```python
#Covert datas to absolute value
#Function to convert data to absolute value
def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '$ {v:d}'.format(v=val)
    return my_format
```

```python
#Plot using Pizza Graphic

#Figure size
plt.figure(figsize = (16,6))

#Pie Chart
plt.pie(result_categorizing_by_segment["Sale_Price"],
        labels = result_categorizing_by_segment["Segment"],
        autopct = autopct_format(result_categorizing_by_segment["Sale_Price"])
       ,startangle = 90)

#Clean inside of the circle
centre_circle = plt.Circle((0,0), 0.80, fc = 'white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

#Labels and anotations
plt.annotate(text = 'Total Sales: ' + '$' + str(int(sum(result_categorizing_by_segment["Sale_Price"]))), 
             xy =(-0.25, 0))
plt.title("Total Sales by Segment")
plt.show()
```

### 7 How many Sales will receive a 15% discount?

### 8 What would be the average sales value before and after the discount?

```python
#All values affected by 15%
before_discount15 = dataset[dataset['Discount'] == 0.15]
before_discount15.head()
```

```python
#Average of all values before the 15% discount
average_before_discount15 = before_discount15.groupby("Discount")['Sale_Price'].mean()
```

```python
print(average_before_discount15)
```

```python
#All value after 15% discount
dataset["Sale_Price_With_Discount"] = dataset["Sale_Price"] - (dataset["Sale_Price"]*dataset["Discount"])
dataset.head()
```

```python
discount15 = dataset[dataset["Discount"] == 0.15]
```

```python
groupby_discount15 = discount15.groupby("Discount")["Sale_Price_With_Discount"]
```

```python
#Average of all values affected by 15
average_after_discount15 = groupby_discount15.mean()
```

```python
print(average_after_discount15)
```

### 9 What is the average sales per segment, per year and per month?

```python
#Create the column of the month
dataset["Month"] = dataset["Order_Date"].dt.month
```

```python
dataset.head()
```

```python
#Group by year, by month and segment,and calculate aggregation statistics
dataset1 = dataset.groupby(['Year', "Month", 'Segment'])["Sale_Price"].agg([np.sum, np.mean, np.median])
```

```python
dataset1
```

```python
#Extract the levels
year = dataset1.index.get_level_values(0)
month = dataset1.index.get_level_values(1)
segment = dataset1.index.get_level_values(2)
```

```python
#Plot
plt.figure(figsize = (12,6))
sns.set()
fig1 = sns.relplot(kind = 'line',
                  data = dataset1,
                  y = 'mean',
                  x = month,
                  hue = segment,
                  col = year,
                  col_wrap =4)
plt.show()
```

```python
#Plot
plt.figure(figsize = (12,6))
sns.set()
fig1 = sns.relplot(kind = 'line',
                  data = dataset1,
                  y = 'sum',
                  x = month,
                  hue = segment,
                  col = year,
                  col_wrap =4)
plt.show()
```

```python
#Plot
plt.figure(figsize = (12,6))
sns.set()
fig1 = sns.relplot(kind = 'line',
                  data = dataset1,
                  y = 'median',
                  x = month,
                  hue = segment,
                  col = year,
                  col_wrap =4)
plt.show()
```

### 10 What The Total Sales By Categories e Subcategories For only Top 12 Subcategories?

```python
dataset.head()
```

```python
#Group by Categorie and SubCategorie, and calculate only numeric sum
dataset2 = dataset.groupby(["Category", "Subcategory"]).sum(numeric_only = True).sort_values("Sale_Price", 
                                                                                               ascending = False).head(12)
```

```python
# Convert the Sales Value column to an integer and sort by category
dataset3 =  dataset2[['Sale_Price']].astype(int).sort_values(by = 'Category').reset_index()
```

```python
# Dataframe with categories and subcategories
dataset3
```

```python
#Create another dataframe with only the totals by category
dataset3_cat = dataset3.groupby('Category').sum(numeric_only = True).reset_index()
```

```python
# Dataframe with categories
dataset3_cat
```

```python
# Color lists for categories
colors_categories = ['#5d00de',
                    '#0ee84f',
                    '#e80e27']
```

```python
# Color lists for subcategories
colors_subcategories = ['#aa8cd4',
                       '#aa8cd5',
                       '#aa8cd6',
                       '#aa8cd7',
                       '#26c957',
                       '#26c958',
                       '#26c959',
                       '#26c960',
                       '#e65e65',
                       '#e65e66',
                       '#e65e67',
                       '#e65e68']
```

```python
# Plot

# Size of the figure
fig, ax = plt.subplots(figsize = (18,12))

# Graphics of the categories
p1 = ax.pie(dataset3_cat['Sale_Price'], 
            radius = 1,
            labels = dataset3_cat['Category'],
            wedgeprops = dict(edgecolor = 'white'),
            colors = colors_categories)

# Graphics of the subcategories
p2 = ax.pie(dataset3['Sale_Price'],
            radius = 0.9,
            labels = dataset3['Subcategory'],
            autopct = autopct_format(dataset3['Sale_Price']),
            colors = colors_subcategories, 
            labeldistance = 0.7,
            wedgeprops = dict(edgecolor = 'white'), 
            pctdistance = 0.53,
            rotatelabels = True)

# Clean the center of the circle
centre_circle = plt.Circle((0, 0), 0.6, fc = 'white')

# Labels and notes
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.annotate(text = 'Total Sales: ' + '$ ' + str(int(sum(dataset3['Sale_Price']))), xy = (-0.2, 0))
plt.title('Total sales by category and Top 12 Subcategories')
plt.show()
```
