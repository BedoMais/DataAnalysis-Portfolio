#### Line Plot

```python
#Line Plot

#Data
x = linspace(0, 5, 10)

y = x**2

#Create the figure
fig = plt.figure()

#Set the scale of the axes
axes = fig.add_axes([0, 0, 0.8, 0.8])

#Create the graph
axes.plot(x, y, 'r')

#Label and Title
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Line Plot')
```

```python
#Line Plot with different scales

#Data
x= linspace(0,5,10)
y= x**2

#Create subplot
fig, axes = plt.subplots(1, 2, figsize=(10,4))

#Create plot1
axes[0].plot(x, x**2, x, exp(x))
axes[0].set_title("Standard Scale")

#Create plot2
axes[1].plot(x, x**2, x, exp(x))
axes[1].set_yscale("log")
axes[1].set_title('Logarithmic Scale(y)')
```

```python
#Grid


#Data
x = linspace(0,5,10)
y= x**2

#Create subplots
fig, axes = plt.subplots(1, 2, figsize = (10,3))


#Standard Grid
axes[0].plot(x, x**2, x, x**3, lw = 2)
axes[0].grid(True)

#Custom Grid
axes[1].plot(x, x**2, x, x**3, lw = 2)
axes[1].grid(color = 'r', alpha = 0.7, linestyle = 'dashed', linewidth = 0.8)

```

```python
#Different styles of pilots

#Data
xx = np.linspace(-0.75, 1, 100)
n = np.array([0,1,2,3,4,5])

#Subplot
fig, axes = plt.subplots(1, 4, figsize = (12,3))

#plot1
axes[0].scatter(xx, xx + 0.25 * randn(len(xx)), color = 'black')
axes[0].set_title('scatter')

#plot2
axes[1].step(n, n**2, lw = 2, color ='blue')
axes[1].set_title('step')

#plot3
axes[2].bar(n, n**2, align = 'center', width = 0.5, alpha = 0.5, color = 'magenta')
axes[2].set_title('bar')

#plot4
axes[3].fill_between(x, x**2, x**3,  alpha = 0.5, color = 'green')
axes[3].set_title('fill_between')
```

#### Histogram with Matplotlib

```python
#Histogram

#Data
n = np.random.randn(100000)

#Create subplots
fig, axes = plt.subplots(1, 2, figsize = (12,4))

#Plot1
axes[0].hist(n)
axes[0].set_title('Standard Histogram')
axes[0].set_xlim((min(n), max(n)))

#plot
axes[1].hist(n, cumulative = True, bins = 50)
axes[1].set_title("Culmulativo Histogram")
axes[1].set_xlim((min(n), max(n)))
```

#### 3D Graph with Matplotlib

```python
#Data
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

#Color map function
def ColorMap(phi_m, phi_p):
    return (+alpha - 2 * np.cos(phi_p)*cos(phi_m) - alpha * np.cos(phi_ext - 2 * phi_p))
    

#More Data
phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X,Y = np.meshgrid(phi_p, phi_m)
Z = ColorMap(X, Y).T
```

```python
import seaborn as sea
```

```python
#Data
datas = sea.load_dataset('tips')
```

```python
datas.head()
```

```python
#The jointplot method creates a 2-variable plot with bivariate and univariate graphs.
sea.jointplot(data = datas, x = 'total_bill', y = 'tip', kind = 'reg')
```

```python
#The lmplot() method creates a plot with data and regression models.
sea.lmplot(data = datas, x = 'total_bill', y = 'tip', col ='smoker' )
```

```python
#Building a Dataframe with Pandas
df = pd.DataFrame()
```
